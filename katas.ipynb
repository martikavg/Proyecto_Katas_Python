KATA 1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def frecuencia_letras(cadena):
    """
    Calcula la frecuencia de cada letra en una cadena de texto, ignorando 
    los espacios.
    
    Argumentos:
        cadena (str): La cadena de texto de la cual se contarán las letras.

    Returns:
        diccionario: Un diccionario donde las claves son las letras y 
        los valores son las veces que aparecen en la cadena (excluyendo 
        espacios).
    """
    frecuencias = {}
    for caracter in cadena:
        # Ignoramos los espacios en blanco tal como lo pide el enunciado
        if caracter != " ":
            # Se pasa a minúsculas para no duplicar letras por ser mayúsculas
            letra = caracter.lower()
            frecuencias[letra] = frecuencias.get(letra, 0) + 1
    return frecuencias

# --- Ejemplo de uso ---
print("=== KATA 1 ===")
texto = "BUENAS Tardes"
diccionario = frecuencia_letras(texto)
print (f'Cadena de texto : {texto}')
print(f'Resultado: {diccionario}')
```

KATA 2. Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función *map()*

```python
def duplicar_valores(lista_numeros):
    """
    Obtiene una nueva lista con el doble de cada valor numérico de la lista 
    original usando map.
    
    Argumentos:
        lista_numeros (List[float]): Lista de números que se desean duplicar.

    Returns:
        List[float]: Una nueva lista donde cada elemento es el doble del 
        original.
    """
    # Usamos map() con una función lambda que multiplica cada elemento por 2
    resultado_map = map(lambda x: x * 2, lista_numeros)
    return list(resultado_map)

# --- Ejemplo de uso ---
print("=== KATA 2 ===")
numeros = [1, 2, 3, 4]
print(f'La lista original es: {numeros}')
print(f'Resultado: {duplicar_valores(numeros)}')
```

KATA 3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def filtrar_por_palabra(palabras, objetivo):
    """
    Filtra una lista de palabras devolviendo únicamente aquellas que contienen
    la palabra objetivo.
    
    Argumentos:
        palabras (List[str]): Lista original de palabras donde se realizará la
                              búsqueda.
        objetivo (str): La subcadena o palabra objetivo que se desea buscar.

    Returns:
        List[str]: Lista con las palabras que contienen el objetivo.
    """
    coincidencias = []
    for palabra in palabras:
        if objetivo in palabra:
            coincidencias.append(palabra)
    return coincidencias

# --- Ejemplo de uso ---
print("=== KATA 3 ===")
lista_palabras = ["divino", "novio", "ovino", "vinagre"]
palabra_objetivo = "vino"
print(f'Lista de palabras = {lista_palabras}')
print(f'La palabra objetivo es = {palabra_objetivo}')
print(f'Resultado: {filtrar_por_palabra(lista_palabras, palabra_objetivo)}')
```

KATA 4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función  *map()*

```python
def diferencia_listas(lista1, lista2):
    """
    Calcula la diferencia elemento a elemento entre dos listas numéricas de 
    igual longitud usando map.
    
    Argumentos:
        lista1 (List[float]): Primera lista numérica (minuendos).
        lista2 (List[float]): Segunda lista numérica (sustraendos).

    Returns:
        List[float]: Una nueva lista con el resultado de restar cada elemento 
        de la lista2 a la lista1.
    
    """
    resultado_map = map(lambda x, y: x - y, lista1, lista2)
    return list(resultado_map)

# --- Ejemplo de uso ---
print("=== KATA 4 ===")
l1 = [10, 20, 30, 40]
l2 = [3, 5, 10, 25]
print(f'La lista 1 es: {l1}')
print(f'La lista 2 es: {l2}')
print(f'Resultado de la diferencia entre las dos listas: {diferencia_listas(l1, l2)}')
```

KATA 5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
def evaluar_rendimiento(lista_numeros, nota_aprobado=5):
    """
    Calcula la media de una lista y determina si alcanza la nota de aprobado.
    
    Argumentos:
        lista_numeros (List[float]): Lista de calificaciones o números para 
                calcular la media.
        nota_aprobado (float): Valor límite opcional para aprobar. Por 
                defecto es 5.

    Returns:
        Tuple[float, str]: Una tupla que contiene la media calculada y el 
        estado ("aprobado" o "suspenso").
    """
    if not lista_numeros:
        return (0.0, "suspenso")
    
    media = sum(lista_numeros) / len(lista_numeros)
    estado = "aprobado" if media >= nota_aprobado else "suspenso"
    return (media, estado)

# --- Ejemplo de uso ---
print("=== KATA 5 ===")
notas = [4.5, 6.0, 5.5, 6.0]
print(f'Las notas obtenidas son: {notas}')
print(f'Resultado de la nota promedio y estado: {evaluar_rendimiento(notas)}')
```

KATA 6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial(n):
    """
    Calcula el factorial de un número entero de manera recursiva.
    Si el número es negativo, muestra un mensaje de aviso y no realiza el 
    cálculo.
    
    Argumentos:
        n (int): El número entero del cual se calcula el factorial.Debe ser 
        no negativo.

    Returns:
        int: El valor del factorial de n. Si el número es negativo, muestra 
        un aviso y retorna 0.
    """
    # Verificación de número negativo
    if n < 0:
        print("Por favor, ingrese un número positivo.")
        return None
        
    # Caso base: si el número es 0 o 1, su factorial es 1
    if n <= 1:
        return 1
    
    # Caso recursivo
    return n * factorial(n - 1)

# --- Ejemplo de uso ---
print("=== KATA 6 ===")
n = 4
print(f'Dado el siguiente número: {n}')
print(f'Su factorial es: {factorial(n)}')
```

KATA 7. Genera una función que convierta una lista de tuplas a una lista de strings. 
Usa la función  *map()*

```python
def tuplas_a_strings(lista_tuplas):
    """
    Convierte una lista de tuplas en una lista de strings usando map().
    
    Argumentos:
        lista_tuplas (list[tuple]): Lista que contiene las tuplas a convertir.
    
    Returns:
        List[str]: Una nueva lista con la representación en string de cada tupla.
    """

    # Usamos map con lambda para convertir cada tupla en string
    return list(map(lambda tupla: " ".join(map(str, tupla)), lista_tuplas))
    

# --- Ejemplo de uso ---
print("=== KATA 7 ===")
datos = [("Juan", 23), (3.14, "Pi"), ("True", 0)]
print(f'Lista de tuplas: {datos}')
print(f'Lista de strings: {tuplas_a_strings(datos)}')
```

KATA 10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una excepción personalizada y maneja el error adecuadamente.

```python
class ListaVaciaError(Exception):
    """Excepción lanzada cuando la lista proporcionada está vacía."""
    pass

def calcular_promedio(numeros):
    """
    Calcula el promedio de una lista. Lanza una excepción si no contiene 
    elementos.
    
    Argumentos:
        numeros (List[float]): Lista de números reales de la cual obtener 
        el promedio.

    Returns:
        float: El promedio calculado de los elementos.
    """
    if not numeros:
        raise ListaVaciaError("No se puede calcular el promedio de una lista vacía.")
    return sum(numeros) / len(numeros)

# Ejemplo de uso y manejo:
print("=== KATA 10 ===")
try:
    print(f'Ejemplo 1:Lista de números')
    print(f'Resultado: {calcular_promedio([10,20,30,40])}')
    # Forzamos el error pasándole una lista vacía
    print(f'Ejemplo 2:Lista vacía')
    print(calcular_promedio([]))
except ListaVaciaError as error:
    print(f'-> Mensaje de la excepción: {error}')
```

KATA 12. Genera una función que al recibir una frase devuelva una lista con la longitud de cada palabra. Usa la función map()

```python
def longitud_palabras(frase):
    """
    Calcula la longitud de cada palabra presente en una frase utilizando map().
    
    Argumentos:
        frase (str): La oración o texto a procesar.

    Returns:
        list[int]: Una lista con los tamaños de cada palabra en el orden original.
    """
    palabras = frase.split()
    resultado_map = map(len, palabras)
    return list(resultado_map)

# --- Ejemplo de uso ---
print("=== KATA 12 ===")
texto = "Katas con Jupyter Notebook"
print(f'Frase : {texto}')
print(f'La longitud de cada palabra de la frase es: {longitud_palabras(texto)}')
```

KATA 13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función  *map()*

```python
def lista_tuplas(caracteres):
    """
    Crea tuplas de (MAYÚSCULA, minúscula) para cada carácter único usando map().
    
    Argumentos:
        caracteres (str o list[str]): Conjunto o colección de caracteres a transformar.

    Returns:
        list[tuple]: Lista de tuplas donde cada una contiene la variante mayúscula y minúscula.
    """
    # 1. Usamos set() para eliminar duplicados automáticamente.
    # 2. Descartamos espacios de forma limpia con un filtro.
    unicos = [c for c in set(caracteres) if c != " "]

    # 3. Aplicamos map() con una lambda
    resultado_map = map(lambda c: (c.upper(), c.lower()), unicos)
    return list(resultado_map)

# --- Ejemplo de uso ---
print("=== KATA 13 ===")
caracteres = "aBbccda"
print(f'Conjunto de caracteres: {caracteres}')
print(f'Resultado lista de tuplas: {lista_tuplas(caracteres)}')
```

KATA 14. Crea una función que retorne las palabras de una lista de palabras que comience con una letra en especifico. Usa la función  *filter()*

```python
def filtrar_por_inicial(palabras, letra_objetivo):
    """
    Filtra una lista de palabras para obtener solo aquellas que inician con una
    letra dada.
    
    Argumentos:
        palabras (list[str]): Lista original de palabras.
        letra_objetivo (str): Carácter inicial buscado.

    Returns:
        list[str]: Palabras que cumplen con la condición (ignora mayúsculas/minúsculas).
    """
    # Pasamos la letra a minúscula una sola vez
    letra = letra_objetivo.lower()
    
    # Filtramos directamente comparando la primera letra [0] de cada palabra en minúsculas
    resultado = filter(lambda p: p.lower()[0] == letra, palabras)
    return list(resultado)

# --- Ejemplo de uso ---
print("=== KATA 14 ===")
lista = ["perro", "Gato", "pájaro", "ratón", "Panda"]
print(f'Lista de palabras: {lista}')
print(f'Palabras que empiezan con "p": {filtrar_por_inicial(lista, 'p')}')
```

```python
def filtrar_palabras_largas(texto, n):
    """
    Usando filter(), extrae las palabras de un texto que superen una longitud 'n'.
    
    Argumentos:
        texto (str): Cadena de texto con las palabras.
        n (int): Cantidad mínima de caracteres requerida.

    Returns:
        list[str]: Lista de palabras cuya longitud es estrictamente mayor que n.
    """
    palabras = texto.split()
    resultado_filter = filter(lambda p: len(p) > n, palabras)
    return list(resultado_filter)

# --- Ejemplo de uso ---
print("=== KATA 16 ===")
texto = "Desarrollando ejercicios de programación avanzada"
n = 8
print(f'Texto : {texto}')
print(f'Las palabras con más de {n} caracteres son: {filtrar_palabras_largas(texto, n)}')
```

KATA 17. Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2] corresponde al número quinientos setenta y dos (572). Usa la función *reduce()*

```python
from functools import reduce

def reconstruir_numero(digitos):
    """
    Construye un número entero a partir de sus dígitos individuales usando reduce().
    
    Params:
        digitos (list[int]): Lista que contiene los dígitos en orden.

    Returns:
        int: El número entero unificado.
    """
    if not digitos:
        return 0
        
    # 1. Convertimos todos los dígitos a cadenas de texto
    cadenas = map(str, digitos)
    
    # 2. Unimos los caracteres con reduce() concatenando texto ("5" + "7" + "2" -> "572")
    numero_str = reduce(lambda acumulado, digito: acumulado + digito, cadenas)
    
    # 3. Convertimos el string resultante a número entero
    return int(numero_str)

# --- Ejemplo de uso ---
print("=== KATA 17 ===")
lista_digitos = [5, 6, 7]
print(f'Lista de digitos : {lista_digitos}')
print(f'Resultado: {reconstruir_numero(lista_digitos)}')
```

```python
def estudiantes_sobresalientes(estudiantes):
    """
    Filtra y extrae aquellos estudiantes que posean una calificación de 90 o 
    superior usando filter().
    
    Params:
        estudiantes (list[dict]): Lista de diccionarios con los datos de cada 
        estudiante.

    Returns:
        list[dict]: Estudiantes sobresalientes con nota mayor o igual a 90.
    """
    # Filtramos directamente evaluando la clave del diccionario
    resultado = filter(lambda estudiante:estudiante["calificación"] >= 90, estudiantes)
    return list(resultado)

# --- Ejemplo de uso ---
print("=== KATA 18 ===")
grupo_estudiantes = [
    {"nombre": "Ana", "edad": 20, "calificación": 85},
    {"nombre": "Luis", "edad": 22, "calificación": 95},
    {"nombre": "María", "edad": 21, "calificación": 90},
    {"nombre": "Carlos", "edad": 19, "calificación": 60}
]
# Para que resultado de estudiantes aparezca en cada línea iteramos
estudiantes_sobresalientes = estudiantes_sobresalientes(grupo_estudiantes)
print("Estudiantes con calificación >= 90:")
for estudiante in estudiantes_sobresalientes:
    print(estudiante)
```

KATA 20. Para una lista con elementos tipo integer y string obtén una nueva lista sólo con los valores int. Usa la función  *filter()*

```python
def extraer_enteros(lista_mixta):
    """
    De una lista heterogénea filtra exclusivamente los elementos de tipo entero 
    usando filter().
    
    Argumentos:
        lista_mixta (list): Lista con datos de naturaleza variada (str, int, etc.).

    Returns:
        list[int]: Lista filtrada que contiene solo los elementos de tipo integer.
    """
    # Usamos type(x) is int , porque discrimina de forma exacta los enteros, 
    # evitando herencias como bool
    
    resultado_filter = filter(lambda x: type(x) is int, lista_mixta)
    return list(resultado_filter)

# --- Ejemplo de uso ---
print("=== KATA 20 ===")
lista_mixta = ["hola", 5, "mundo", 12, True, 42, "python"]
print(f'Lista mixta : {lista_mixta}')
print(f'Los elementos enteros de la lista son: {extraer_enteros(lista_mixta)}')
```

KATA 22. Dada una lista numérica, obtén el producto total de los valores de dicha lista. Usa la función *reduce()*.

```python
def producto_total(lista_numeros):
    """
    Multiplica todos los números de una lista para obtener su producto acumulado.
    
    Params:
        lista_numeros (list[int | float]): Lista de números a multiplicar.

    Returns:
        int | float: El resultado de la multiplicación de todos los elementos.
    """
    if not lista_numeros:
        return 0
        
    return reduce(lambda x, y: x * y, lista_numeros)

# --- Ejemplo de uso ---
print("=== KATA 22 ===")
numeros = [2, 3, 4, 5]
print(f'El producto total de {numeros} es: {producto_total(numeros)}')
```

KATA 23. Concatena una lista de palabras. Usa la función *reduce()*.

```python
def concatenar_palabras(lista_palabras, separador=" "):
    """
    Concatena una lista de cadenas de texto en una sola frase usando reduce().
    
    Params:
        lista_palabras (list[str]): Lista de palabras a concatenar.
        separador (str): Carácter o espacio para unir las palabras (por defecto " ").

    Returns:
        str: Cadena de texto resultante.
    """
    if not lista_palabras:
        return ""
        
    return reduce(lambda x, y: f'{x}{separador}{y}', lista_palabras)

# --- Ejemplo de uso ---
print("=== KATA 23 ===")
palabras = ["La", "práctica", "ayuda", "a", "aprender"]
print(f'Lista de palabras : {palabras}')
print(f'Resultado concatenado de palabras: "{concatenar_palabras(palabras)}"')
```

KATA 24. Calcula la diferencia total en los valores de una lista. Usa la función *reduce()*.

```python
def diferencia_total(lista_numeros):
    """
    Resta secuencialmente todos los elementos de la lista empezando por el primero.
    
    Params:
        lista_numeros (list[int | float]): Lista de números a restar.

    Returns:
        int | float: El resultado de restar cada elemento acumulativamente.
    """
    if not lista_numeros:
        return 0
        
    return reduce(lambda x, y: x - y, lista_numeros)

# --- Ejemplo de uso ---
print("=== KATA 24 ===")
valores = [80, 20, 10, 5]
# Operación realizada: (80 - 20 - 10 - 5) = 45
print(f'La diferencia acumulada de {valores} es: {diferencia_total(valores)}')
```

KATA 26. Crea una función  *lambda*  que calcule el resto de la división entre dos números dados.

```python
def calcular_resto(dividendo, divisor):
    """
    Calcula el resto de la división entre dos números usando una función lambda.
    
    Params:
        dividendo (int | float): Número que se va a dividir.
        divisor (int | float): Número por el cual se divide.

    Returns:
        int | float: El resto (módulo) de la división.
    """
    resto = lambda a, b: a % b

    try:
        return resto(dividendo, divisor)

    # Establecemos una excepción cuando se divide entre cero
    except ZeroDivisionError:
        print("[!] Error: No es posible dividir entre cero.")
        return None

# --- Ejemplo de uso ---
print("=== KATA 26 ===")
print(f'Ejemplo 1:')
print(f'Resto de 10 % 3: {calcular_resto(10, 3)}')
print(f'Ejemplo 2:')
print(f'Resto de 15 % 0: {calcular_resto(15, 0)}')
```

KATA 27. Crea una función que calcule el promedio de una lista de números.

KATA 28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

```python
def primer_duplicado(lista_elementos):
    """
    Encuentra el primer elemento que aparece repetido al recorrer la lista de 
    izquierda a derecha.
    
    Params:
        lista_elementos (list): Lista con elementos de cualquier tipo.

    Returns:
        object | None: El primer elemento repetido si existe, de lo contrario None.
    """
    # Usamos un conjunto (set) para almacenar los elementos que ya hemos visto.
    vistos = set()
    
    for elemento in lista_elementos:
        # Si el elemento ya está en nuestro registro de vistos, ¡es el primer duplicado!
        if elemento in vistos:
            return elemento
        
        # Si no lo habíamos visto antes, lo añadimos al conjunto de vistos.
        vistos.add(elemento)
        
    # Si termina el bucle sin encontrar repetidos, devolvemos None
    return None

# --- Ejemplo de uso ---
print("=== KATA 28 ===")
datos = ["manzana", "banana", "pera", "banana", "manzana"]
print(f'Lista dada : {datos}')
resultado = primer_duplicado(datos)

if resultado is not None:
    print(f'El primer elemento duplicado es: "{resultado}"')
else:
    print("No se encontraron elementos duplicados")
```

KATA 29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el carácter '#', excepto los últimos cuatro.

```python
def enmascarar_dato(variable):
    """
    Convierte cualquier tipo de variable a texto y enmascara sus caracteres con '#', 
    conservando únicamente visibles los últimos cuatro.
    
    Params:
        variable (object): Variable de cualquier tipo (int, str, float, etc.).

    Returns:
        str: Cadena formateada y enmascarada.
    """
    # 1. Convertimos cualquier tipo de dato a cadena de texto
    texto = str(variable)
    
    # 2. Si el texto tiene 4 caracteres o menos, no hay nada que ocultar
    if len(texto) <= 4:
        return texto
        
    # 3. Construimos la mascara con '#' para los caracteres a ocultar
    caracteres_ocultos = "#" * (len(texto) - 4)

    # 4. Extraemos los últimos 4 caracteres (slicing)
    ultimos_cuatro = texto[-4:]
    
    # 5. Unimos ambas partes y devolvemos el resultado
    return caracteres_ocultos + ultimos_cuatro

# --- Ejemplo de uso ---
print("=== KATA 29 ===")
tarjeta_credito = 1234567890123456
usuario = "Miguel2134"
clave_corta = 982

print(f'Tarjeta enmascarada: {enmascarar_dato(tarjeta_credito)}')
print(f'Usuario enmascarado: {enmascarar_dato(usuario)}')
print(f'Clave corta:         {enmascarar_dato(clave_corta)}')
```

KATA 30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

```python
def es_anagrama(palabra1, palabra2):
    """
    Comprueba si dos palabras son anagramas comparando sus letras ordenadas.
    
    Argumentos:
        palabra1 (str): Primera palabra a evaluar.
        palabra2 (str): Segunda palabra a evaluar.

    Returns:
        bool: True si son anagramas, False en caso contrario.
    """
    # Limpiamos espacios y convertimos a minúsculas
    p1 = palabra1.lower().replace(" ", "")
    p2 = palabra2.lower().replace(" ", "")
    
    # Si ordenamos las letras de ambas y coinciden, son anagramas (resultado booleano)
    return sorted(p1) == sorted(p2)

# --- Ejemplo de uso ---
print("=== KATA 30 ===")
print(f"¿'Roma' y 'Amor' son anagramas?: {es_anagrama('Roma', 'Amor')}")
print(f"¿'Hola' y 'Mundo' son anagramas?: {es_anagrama('Hola', 'Mundo')}")
```

KATA 32. Crea una función que tome un nombre completo y una lista de empleados, busque el nombre completo en la lista y devuelve el puesto del empleado si está en la lista, de lo contrario, devuelve un mensaje indicando que la persona no trabaja aquí.

```python
def puesto_empleado(nombre_completo, empleados):
    """
    Busca el puesto de trabajo de un empleado dentro de una lista de empleados.
    
    Params:
        nombre_completo (str): Nombre y apellido del empleado buscado.
        empleados (list[dict]): Lista de diccionarios con claves 'nombre' y 'puesto'.

    Returns:
        str: El puesto del empleado o un mensaje indicando que no pertenece a la empresa.
    """

    # Normalizamos el nombre para que en la comparación no aparezcan detalles (como 
    # espacios de más o diferencias entre mayúsculas y minúsculas) que rompan la búsqueda.
    
    nombre_normalizado = nombre_completo.strip().lower()
    
    for emp in empleados:
        if emp.get("nombre", "").strip().lower() == nombre_normalizado:
            return f"Puesto de {emp['nombre']}: {emp['puesto']}"
            
    return f"La persona '{nombre_completo}' no trabaja aquí."

# --- Ejemplo de uso ---
print("=== KATA 32 ===")
registro_empleados = [
    {"nombre": "Ana Gómez", "puesto": "Desarrolladora Python"},
    {"nombre": "Carlos Ruiz", "puesto": "Analista de Datos"},
    {"nombre": "Lucía Martín", "puesto": "Diseñadora UX"}
]

print(puesto_empleado("carlos ruiz", registro_empleados))
print(puesto_empleado("Pedro Picapiedra", registro_empleados))
```

KATA 34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son: *crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol* .
>> Código a seguir:
> 1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
> 2. Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
> 3. Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
> 4. Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
> 5. Implementar el método quitar_rama para eliminar una rama en una posición específica.
> 6. Implementar el método info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las mismas.

>>Caso de uso:
> 1. Crear un árbol.
> 2. Hacer crecer el tronco del árbol una unidad.
> 3. Añadir una nueva rama al árbol.
> 4. Hacer crecer todas las ramas del árbol una unidad.
> 5. Añadir dos nuevas ramas al árbol.
> 6. Retirar la rama situada en la posición 2.
> 7. Obtener información sobre el árbol.

```python
class Arbol:
    """
    Representa la estructura de un árbol con tronco y ramas independientes.
    """
    # ---------------------------------------------------------------------
    # CONSIGNA 1: Inicializar un árbol con un tronco de longitud 1 
    # y una lista vacía de ramas.
    # ---------------------------------------------------------------------
    def __init__(self):
        """Inicializa el árbol con tronco de longitud 1 y sin ramas."""
        self.tronco = 1
        self.ramas = []

    # ---------------------------------------------------------------------
    # CONSIGNA 2: Implementar el método crecer_tronco para aumentar 
    # la longitud del tronco en una unidad.
    # ---------------------------------------------------------------------
    def crecer_tronco(self):
        """Aumenta la longitud del tronco en una unidad."""
        self.tronco += 1
        print(f"El tronco ha crecido. Longitud actual del tronco: {self.tronco}")

    # ---------------------------------------------------------------------
    # CONSIGNA 3: Implementar el método nueva_rama para agregar 
    # una nueva rama de longitud 1 a la lista de ramas.
    # ---------------------------------------------------------------------    
    def nueva_rama(self):
        """Agrega una nueva rama de longitud 1 a la lista."""
        self.ramas.append(1)
        print(f"Se ha añadido una nueva rama (Longitud: 1). Total de ramas: {len(self.ramas)}")

    # ---------------------------------------------------------------------
    # CONSIGNA 4: Implementar el método crecer_ramas para aumentar 
    # en una unidad la longitud de todas las ramas existentes.
    # ---------------------------------------------------------------------
    def crecer_ramas(self):
        """Aumenta en una unidad la longitud de todas las ramas existentes."""
        self.ramas = [longitud + 1 for longitud in self.ramas]

    # ---------------------------------------------------------------------
    # CONSIGNA 5: Implementar el método quitar_rama para eliminar 
    # una rama en una posición específica.
    # ---------------------------------------------------------------------
    def quitar_rama(self, posicion):
        """
        Elimina la rama ubicada en el índice/posición especificado.

        Params:
            posicion (int): Índice de la rama a eliminar (inicia en 0).
        """

        try:
            if 0 <= posicion < len(self.ramas):
                rama_eliminada = self.ramas.pop(posicion)
                print(f"Se ha quitado la rama en posición {posicion} (longitud era {rama_eliminada}).")
            else:
                raise IndexError(f"No existe ninguna rama en la posición {posicion}.")
        except IndexError as error:
            print(f"[!] Error: {error}")

    # ---------------------------------------------------------------------
    # CONSIGNA 6: Implementar el método info_arbol para devolver 
    # información sobre la longitud del tronco, el número de ramas 
    # y las longitudes de las mismas.
    # ---------------------------------------------------------------------
    def info_arbol(self):
        """
        Devuelve información detallada de la estructura actual del árbol.

        Returns:
            dict: Información sobre el tronco, cantidad de ramas y sus longitudes.
        """
        return {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitudes_ramas": self.ramas
        }
```

KATA 36. Crea la clase *UsuarioBanco* ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta
corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y
agregar dinero al saldo.

>> Código a seguir:
> 1. Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante True y False .
> 2. Implementar el método retirar_dinero para retirar dinero del saldo del usuario. Lanzará un error en caso de no
poder hacerse.
> 3. Implementar el método transferir_dinero para realizar una transferencia desde otro usuario al usuario actual.
Lanzará un error en caso de no poder hacerse.
> 4. Implementar el método agregar_dinero para agregar dinero al saldo del usuario.

>> Caso de uso:
> 1. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
PROYECTO LÓGICA: Katas de Python 3
> 2. Agregar 20 unidades de saldo de "Bob".
> 3. Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
> 4. Retirar 50 unidades de saldo a "Alicia".

```python
class UsuarioBanco:
    """
    Representa a un usuario de un banco con su nombre, saldo disponible
    y el estado de su cuenta corriente.
    """

    # ---------------------------------------------------------------------
    # CONSIGNA 1: Inicializar un usuario con su nombre, saldo y si tiene 
    # o no cuenta corriente mediante True y False.
    # ---------------------------------------------------------------------
    def __init__(self, nombre, saldo, cuenta_corriente):
        """
        Inicializa una nueva instancia de usuario de banco.

        Params:
            nombre (str): Nombre del titular de la cuenta.
            saldo (int | float): Saldo inicial en la cuenta.
            cuenta_corriente (bool): True si posee cuenta corriente activa, False en 
                                     caso contrario.
        """
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    # ---------------------------------------------------------------------
    # CONSIGNA 2: Implementar el método retirar_dinero para retirar dinero 
    # del saldo del usuario. Lanzará un error en caso de no poder hacerse.
    # ---------------------------------------------------------------------
    def retirar_dinero(self, cantidad):
        """
        Retira una cantidad del saldo del usuario siempre que disponga de fondos 
        suficientes.

        Params:
            cantidad (int | float): Monto a retirar.
        """
        try:
            if cantidad <= 0:
                raise ValueError("La cantidad a retirar debe ser mayor a cero.")
            if cantidad > self.saldo:
                raise ValueError(f"{self.nombre} no tiene saldo suficiente para retirar {cantidad}. Saldo actual: {self.saldo}")
            
            self.saldo -= cantidad
            print(f"[-] Se han retirado {cantidad} unidades de la cuenta de {self.nombre}. Saldo restante: {self.saldo}")
        except ValueError as error:
            print(f"[!] Error al retirar dinero: {error}")

    # ---------------------------------------------------------------------
    # CONSIGNA 3: Implementar el método transferir_dinero para realizar una 
    # transferencia desde otro usuario al usuario actual. Lanzará un error 
    # en caso de no poder hacerse.
    # ---------------------------------------------------------------------
    def transferir_dinero(self, otro_usuario, cantidad):
        """
        Realiza una transferencia de dinero desde otro usuario hacia el usuario actual.

        Params:
            otro_usuario (UsuarioBanco): Usuario emisor que transfiere el dinero.
            cantidad (int | float): Monto a transferir.
        """
        # Hacemos una comprobación de seguridad para asegurarnos de que el objeto que recibe la función es del tipo correcto.
        try:
            if not isinstance(otro_usuario, UsuarioBanco):
                raise TypeError("El emisor debe ser una instancia de UsuarioBanco.")
            if cantidad <= 0:
                raise ValueError("La cantidad a transferir debe ser mayor a cero.")
            if cantidad > otro_usuario.saldo:
                raise ValueError(f"{otro_usuario.nombre} no tiene saldo suficiente ({otro_usuario.saldo}) para transferir {cantidad}.")

            otro_usuario.saldo -= cantidad
            self.saldo += cantidad
            print(f"[⇄] Transferencia exitosa: {otro_usuario.nombre} ha transferido {cantidad} unidades a {self.nombre}.")
        except (ValueError, TypeError) as error:
            print(f"[!] Error en la transferencia: {error}")

    # ---------------------------------------------------------------------
    # CONSIGNA 4: Implementar el método agregar_dinero para agregar 
    # dinero al saldo del usuario.
    # ---------------------------------------------------------------------
    def agregar_dinero(self, cantidad):
        """
        Agrega una cantidad especificada al saldo actual del usuario.

        Params:
            cantidad (int | float): Monto a ingresar a la cuenta.
        """
        try:
            if cantidad <= 0:
                raise ValueError("La cantidad a agregar debe ser mayor a cero.")
            self.saldo += cantidad
            print(f"[+] Se han agregado {cantidad} unidades a la cuenta de {self.nombre}. Saldo actual: {self.saldo}")
        except ValueError as error:
            print(f"[!] Error al agregar dinero: {error}")
```

37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras , reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función procesar_texto .

> > Código a seguir:
> 1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
> 2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva . Tiene que devolver el texto con el remplazo de palabras.
> 3. Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
> 4. Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.

>> Caso de uso:
> Comprueba el funcionamiento completo de la función procesar_texto

```python
# Crear una función llamada "Procesar_texto"

# ---------------------------------------------------------------------
# CONSIGNA 1: Crear una función contar_palabras para contar el número 
# de veces que aparece cada palabra en el texto. Tiene que devolver un 
# diccionario.
# ---------------------------------------------------------------------
def contar_palabras(texto):
    """
    Cuenta la frecuencia de cada palabra dentro de una cadena de texto.

    Params:
        texto (str): El texto a analizar.

    Returns:
        dict: Diccionario con las palabras como claves y sus frecuencias como valores.
    """
    palabras = texto.lower().replace(".", "").replace(",", "").split()
    conteo = {}
    for palabra in palabras:
        conteo[palabra] = conteo.get(palabra, 0) + 1
    return conteo

# ---------------------------------------------------------------------
# CONSIGNA 2: Crear una función reemplazar_palabras para reemplazar 
# una palabra_original del texto por una palabra_nueva. Tiene que devolver 
# el texto con el reemplazo de palabras.
# ---------------------------------------------------------------------
def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """
    Reemplaza todas las ocurrencias de una palabra original por una nueva.

    Params:
        texto (str): El texto base.
        palabra_original (str): Palabra a buscar.
        palabra_nueva (str): Palabra con la que se reemplazará.

    Returns:
        str: Texto resultante tras la sustitución.
    """
    return texto.replace(palabra_original, palabra_nueva)

# ---------------------------------------------------------------------
# CONSIGNA 3: Crear una función eliminar_palabra para eliminar 
# una palabra del texto. Tiene que devolver el texto con la palabra 
# eliminada.
# ---------------------------------------------------------------------
def eliminar_palabra(texto, palabra):
    """
    Elimina todas las ocurrencias de una palabra en el texto.

    Params:
        texto (str): El texto base.
        palabra (str): La palabra a eliminar.

    Returns:
        str: Texto resultante tras remover la palabra.
    """
    palabras = texto.split()
    palabras_filtradas = [p for p in palabras if p.strip(".,!?") != palabra]
    return " ".join(palabras_filtradas)

# ---------------------------------------------------------------------
# CONSIGNA 4: Crear la función procesar_texto que tome un texto, una 
# opción ("contar", "reemplazar", "eliminar") y un número de argumentos 
# variables según la opción indicada.
# ---------------------------------------------------------------------
def procesar_texto(texto, opcion, *args):
    """
    Procesa un texto según la opción especificada llamando a la función correspondiente.

    Params:
        texto (str): El texto a procesar.
        opcion (str): Opción a ejecutar ('contar', 'reemplazar', 'eliminar').
        *args: Argumentos variables según la opción indicada.

    Returns:
        dict | str | None: El resultado del procesamiento o None si ocurre un error.
    """
    try:
        if not isinstance(texto, str):
            raise TypeError("El texto ingresado debe ser una cadena de caracteres (str).")

        if opcion == "contar":
            print("Ejecutando función: contar_palabras...")
            return contar_palabras(texto)

        elif opcion == "reemplazar":
            if len(args) < 2:
                raise ValueError("La opción 'reemplazar' requiere dos argumentos: (palabra_original, palabra_nueva).")
            print(f"Ejecutando función: reemplazar_palabras ('{args[0]}' -> '{args[1]}')...")
            return reemplazar_palabras(texto, args[0], args[1])

        elif opcion == "eliminar":
            if len(args) < 1:
                raise ValueError("La opción 'eliminar' requiere un argumento: (palabra_a_eliminar).")
            print(f"Ejecutando función: eliminar_palabra ('{args[0]}')...")
            return eliminar_palabra(texto, args[0])

        else:
            raise ValueError(f"Opción inválida: '{opcion}'. Las opciones válidas son: 'contar', 'reemplazar', 'eliminar'.")

    except (TypeError, ValueError) as error:
        print(f"[!] Error al procesar texto: {error}")
        return None
```

KATA 38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.

```python
from datetime import datetime, time

def determinar_momento_dia(hora_str):
    """
    Determina si es de día, tarde o noche según la hora proporcionada.

    Params:
        hora_str (str): La hora ingresada en formato HH:MM (hora:minutos) de 24horas.

    Returns:
        str: El momento del día correspondiente ("Día", "Tarde", "Noche").
    """
    try:
        # 1. Convertimos la cadena de texto a un objeto time de Python
        hora_objeto = datetime.strptime(hora_str, "%H:%M").time()

        # 2. Definimos los límites para cada rango
        inicio_dia = time(0, 0)
        inicio_tarde = time(13, 0)
        inicio_noche = time(20, 0)

        # 3. Evaluamos según el objeto time
        if inicio_dia <= hora_objeto < inicio_tarde:
            return "Día"
        elif inicio_tarde <= hora_objeto < inicio_noche:
            return "Tarde"
        else:
            return "Noche"
    # 4. Devuelve error sino cumple con formato y tiempo
    except ValueError:
        return "Error: Ingresar en formato 'HH:MM'(hora 00-23 y minutos 00-60)."


# --- Ejemplo de uso ---
print("=== KATA 38 (con datetime) ===")
horas_prueba = ["8", "15:45", "22:10", "03:00", "24:00"]

for h in horas_prueba:
    print(f"Hora: {h} -> Resultado: {determinar_momento_dia(h)}")
```

KATA 39. Escribe un programa que determine qué calificación en texto tiene un alumno en base a su calificación numérica.
>> Las reglas de calificación son:
> - 0 - 69 insuficiente
> - 70 - 79 bien
> - 80 - 89 muy bien
> - 90 - 100 excelente

```python
def determinar_calificacion(nota):
    """
    Determina la calificación en texto de un alumno en base a su nota numérica.

    Params:
        nota (int | float): La nota numérica del alumno (0 a 100).

    Returns:
        str: La calificación en texto ("insuficiente", "bien", "muy bien", "excelente").
    """
    # 1. Validamos que la nota sea un valor numérico y esté en el rango de 0 a 100
    if not isinstance(nota, (int, float)) or nota < 0 or nota > 100:
        return "Error: La nota debe ser un número entre 0 y 100."

    # 2. Evaluamos los rangos de calificación
    if nota <= 69:
        return "insuficiente"
    elif nota <= 79:
        return "bien"
    elif nota <= 89:
        return "muy bien"
    else:
        return "excelente"


# --- Ejemplo de uso ---
print("=== KATA 39 ===")
notas_prueba = [55, 75, 84, 98, 105, -5]

for n in notas_prueba:
    print(f"Nota: {n} -> Calificación: {determinar_calificacion(n)}")
```

KATA 40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math

def calcular_area(figura, datos):
    """
    Calcula el área de una figura geométrica en función del tipo y sus datos.

    Params:
        figura (str): Nombre de la figura ("rectangulo", "circulo" o "triangulo").
        datos (tuple): Tupla con los datos requeridos:
                       - "rectangulo": (base, altura)
                       - "circulo": (radio,)
                       - "triangulo": (base, altura)

    Returns:
        float | int | str: El área calculada o un mensaje de error si los datos son 
        inválidos.
    """
    try:
        # 1. Normalizamos el texto de la figura
        figura_limpia = str(figura).lower().strip()

        # 2. Evaluamos según la figura y desempaquetamos la tupla datos
        if figura_limpia == "rectangulo":
            base, altura = datos
            return base * altura

        elif figura_limpia == "circulo":
            (radio,) = datos
            return math.pi * (radio ** 2)

        elif figura_limpia == "triangulo":
            base, altura = datos
            return (base * altura) / 2

        else:
            return f"Error: Figura '{figura}' no reconocida. Usar: 'rectangulo', 'circulo' o 'triangulo'."

    except (ValueError, TypeError):
        return f"Error: Datos {datos} no válidos para calcular el área de un {figura}."


# --- Ejemplo de uso ---
print("=== KATA 40 ===")

pruebas = [
    ("rectangulo", (10, 5)),
    ("circulo", (4,)),
    ("triangulo", (6, 8)),
    ("pentagono", (5, 3))
]

# Hacemos bucle for y desempaquetado de tuplas para mostrar resultados 
for fig, dts in pruebas:
    resultado = calcular_area(fig, dts)
    # Validamos el tipo de resultado
    if isinstance(resultado, (int, float)):
        # Alineamos a la izquierda , indicamos caracteres y decimales en resultado
        print(f"Figura: {fig:<10} | Datos: {str(dts):<8} -> Área: {resultado:.2f}")
    else:
        print(f"Figura: {fig:<10} | Datos: {str(dts):<8} -> {resultado}")
```

KATA 41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe hacer lo siguiente:
> 1. Solicita al usuario que ingrese el precio original de un artículo.
> 2. Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).
> 3. Si el usuario responde que sí, solicita que ingrese el valor del cupón de > > > descuento.
> 4. Aplica el descuento al precio original del artículo, siempre y cuando el valor > del cupón sea válido (es decir, mayor
> a cero). Por ejemplo, descuento de 15€.
> 5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado > o sin él.
> 6. Recuerda utilizar estructuras de control de flujo como if, elif y else para > llevar a cabo estas acciones en tu programa de Python.

```python
def calcular_monto_final(precio_original, tiene_cupon, valor_cupon):
    """
    Calcula el monto final de una compra después de aplicar un descuento.

    Params:
        precio_original (int | float): El precio base del artículo.
        tiene_cupon (str): Respuesta del usuario ('sí', 'si', 'no').
        valor_cupon (int | float, optional): El valor a descontar en la compra. Default es 0.

    Returns:
        float | str: El monto final a pagar o un mensaje de error si los datos son inválidos.
    """
    
    try:
        precio = float(precio_original)
        if precio <= 0:
            return "Error: El precio original debe ser un número mayor a cero."

        respuesta = str(tiene_cupon).strip().lower()

        # Evaluación de cupón de descuento
        if respuesta in ["sí", "si"]:
            descuento = float(valor_cupon)
            if descuento > 0:
                # Aplica el descuento si es válido
                monto_final = max(0.0, precio - descuento)
            else:
                # Cupón inválido (menor o igual a cero), no se aplica descuento
                monto_final = precio
        elif respuesta == "no":
            monto_final = precio
        else:
            return "Error: Respuesta no válida para el cupón. Indique 'sí' o 'no'."

        # Devuelve el precio final calculado
        return monto_final

    except (ValueError, TypeError):
        return "Error: Ingrese valores numéricos válidos para los precios y descuentos."
```

```python
# --- Demostración del programa ---
print("=== KATA 41 ===")

# Simulaciones de prueba (casos posibles)
casos_prueba = [
    (100.0, "sí", 15.0),   # Caso válido con descuento
    (50.0, "no", 0),       # Caso válido sin descuento
    (80.0, "si", -5.0),    # Cupón menor o igual a cero
    ("abc", "no", 0),      # Error: precio inválido
    (100.0, "quizás", 10)  # Error: respuesta de cupón inválida
]

for precio, tiene_c, val_c in casos_prueba:
    print("\n----------------------------------")
    # CONSIGNA 1: Muestra siempre el precio original ingresado
    print(f"1. Precio original ingresado: {precio}€")

    # CONSIGNA 2: Muestra siempre la respuesta del cupón
    print(f"2. ¿Tiene cupón de descuento?: {tiene_c}")

    # CONSIGNA 3: Si la respuesta fue 'sí'/'si', muestra el valor del cupón ingresado
    if str(tiene_c).strip().lower() in ["sí", "si"]:
        print(f"3. Valor del cupón ingresado: {val_c}€")

    # Ejecución del cálculo
    resultado = calcular_monto_final(precio, tiene_c, val_c)

    # CONSIGNA 5: Muestra el precio final o el mensaje de error explícito
    if isinstance(resultado, (int, float)):
        print(f"5. Precio final de la compra: {resultado:.2f}€")
    else:
        print(f"5. Resultado: {resultado}")
```

## 1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def frecuencia_letras(cadena) -> dict[str, int]:
    """
    Calcula la frecuencia de cada letra en una cadena de texto, ignorando los espacios.
    
    Argumentos:
        cadena (str): La cadena de texto de la cual se contarán las letras.
    """
    frecuencias = {}
    for caracter in cadena:
        if caracter != " ":
            # Se suele pasar a minúsculas para no duplicar letras por mayúsculas
            letra = caracter.lower()
            frecuencias[letra] = frecuencias.get(letra, 0) + 1
    return frecuencias

# --- Ejemplo de uso ---
resultado = frecuencia_letras("Buenas Tardes")
print(resultado)
```

## 2. Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función *map()*

```python
def duplicar_valores(lista):
    """
    Multiplica por dos cada valor de la lista.
    
    Argumentos:
        lista (float): Lista original de números a duplicar.
    """
    # Usamos map() con una función lambda que multiplica cada elemento por 2
    return list(map(lambda x: x * 2, lista))

# --- Ejemplo de uso ---
numeros = [1, 2, 3, 4]
print(duplicar_valores(numeros))
```

## 3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
def filtrar_por_palabra(palabras, objetivo) -> list[str]:
    """
    Filtra una lista de palabras buscando aquellas que contengan la palabra objetivo.
    
    Argumentos:
        palabras (list): Lista de palabras en la que buscar.
        objetivo (str): La palabra o subcadena que se busca encontrar dentro de los elementos.
    """
    return [palabra for palabra in palabras if objetivo in palabra]

# --- Ejemplo de uso ---
lista_palabras = ["divino", "novio", "ovino", "vinagre"]
palabra_objetivo = "vino"
print(filtrar_por_palabra(lista_palabras, palabra_objetivo))
```

## 4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función *map()*

```python
def diferencia_listas(lista1, lista2):
    """
    Calcula la diferencia elemento a elemento de dos listas numéricas usando map().
    
    Argumentos:
        lista1 (float): Primera lista de números.
        lista2 (float): Segunda lista de números.
    """
    return list(map(lambda x, y: x - y, lista1, lista2))

# --- Ejemplo de uso ---
l1 = [10, 20, 30]
l2 = [2, 5, 10]
print(diferencia_listas(l1, l2))
```

## 5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
def evaluar_rendimiento(lista_numeros, nota_aprobado=5) -> tuple[float, str]:
    """
    Calcula la media de una lista de números y evalúa si supera la nota de aprobado.
    
    Argumentos:
        lista_numeros (float): Lista de calificaciones numéricas.
        nota_aprobado (float): Calificación mínima requerida para aprobar (por defecto es 5).
    """
    if not lista_numeros:
        return (0.0, "suspenso")
    
    media = sum(lista_numeros) / len(lista_numeros)
    estado = "aprobado" if media >= nota_aprobado else "suspenso"
    return (media, estado)

# --- Ejemplo de uso ---
notas = [4.5, 6.0, 5.5, 6.0]
print(evaluar_rendimiento(notas))
```

## 6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial(n):
    """
    Calcula el factorial de un número entero de manera recursiva.
    Si el número es negativo, muestra un mensaje de aviso y no realiza el cálculo.
    
    Argumentos:
        n (int): El número entero del cual se calculará el factorial.
    """
    # Verificación de número negativo
    if n < 0:
        print("Por favor, ingrese un número positivo.")
        return None
        
    # Caso base: si el número es 0 o 1, su factorial es 1
    if n <= 1:
        return 1
    
    # Caso recursivo
    return n * factorial(n - 1)

# --- Ejemplo de uso ---
print(factorial(5))
```

## 7. Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función *map()*

```python
def tuplas_a_strings(lista_tuplas):
    """
    Convierte una lista de tuplas en una lista de strings usando map() y 
    el método join.
    
    Argumentos:
        lista_tuplas: list[tuple]) -> list[str]: Lista que contiene las tuplas 
        a convertir.
    """
    # map(str, tupla) convierte cada elemento de una tupla a texto.
    # Primero usamos una función auxiliar para unir esos elementos en un único string.
    def unir_tupla(tupla):
        return "".join(map(str, tupla))
        
    # Aplicamos map() para procesar toda la lista usando la función auxiliar
    return list(map(unir_tupla, lista_tuplas))

# --- Ejemplo de uso ---
datos = [('a', 'b'), ('c', 'd'), (1, 2)]
print(tuplas_a_strings(datos))
```
